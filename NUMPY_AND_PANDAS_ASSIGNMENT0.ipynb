```python
import os
import pandas as pd

# The 'path' variable holds the directory where the dataset files were extracted.
# We assume the main data file is 'FIFA23_official_data.csv'.
csv_file_path = os.path.join(path, 'FIFA23_official_data.csv')

df = pd.read_csv(csv_file_path)

print("DataFrame loaded successfully. Displaying the first 5 rows:")
display(df.head())
```

```python
display(df.describe())
```

### Handling Missing Values

```python
# Display the count of missing values for each column
print("Missing values before handling:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

#### Using `.fillna()`

```python
# Fill missing values in a numerical column ('Kit Number') with its mean
df['Kit Number'].fillna(df['Kit Number'].mean(), inplace=True)
print("Missing values in 'Kit Number' after fillna with mean:")
display(df['Kit Number'].isnull().sum())

# Fill missing values in an object/string column ('Loaned From') with a placeholder
df['Loaned From'].fillna('Never Loaned', inplace=True)
print("Missing values in 'Loaned From' after fillna with 'Never Loaned':")
display(df['Loaned From'].isnull().sum())
```

#### Using `.dropna()`

```python
# Create a copy of the DataFrame to demonstrate dropna without altering the original 'df' too much
df_dropped = df.copy()

# Drop rows that have any missing values
before_drop = df_dropped.shape[0]
df_dropped.dropna(inplace=True)
after_drop = df_dropped.shape[0]

print(f"Number of rows before dropna: {before_drop}")
print(f"Number of rows after dropna: {after_drop}")
print(f"Rows dropped: {before_drop - after_drop}")

print("\nMissing values after dropping rows with any NaNs:")
display(df_dropped.isnull().sum()[df_dropped.isnull().sum() > 0])
```

### Grouping and Aggregating Data

```python
# Group by 'Club' and calculate the mean 'Overall' rating
club_overall_rating = df.groupby('Club')['Overall'].mean().sort_values(ascending=False)
print("Average Overall rating by Club (top 10):")
display(club_overall_rating.head(10))
```

```python
# Group by 'Nationality' and count the number of players
nationality_counts = df.groupby('Nationality')['ID'].count().sort_values(ascending=False)
print("Number of players by Nationality (top 10):")
display(nationality_counts.head(10))
```

### Selecting Columns

```python
# Select multiple columns
selected_columns = df[['Name', 'Age', 'Overall', 'Club']]
print("Selected 'Name', 'Age', 'Overall', and 'Club' columns (first 5 entries):")
display(selected_columns.head())
```

### Filtering Rows

```python
# Filter rows where 'Age' is greater than 30
experienced_players = df[df['Age'] > 30]
print("Players older than 30 (first 5 entries):")
display(experienced_players.head())
```

### Applying Functions

```python
# Define a simple function to categorize players by age
def age_category(age):
    if age < 20:
        return 'Young'
    elif 20 <= age <= 28:
        return 'Prime'
    else:
        return 'Veteran'

# Apply the function to create a new 'Age_Category' column
df['Age_Category'] = df['Age'].apply(age_category)
print("DataFrame with new 'Age_Category' column (first 5 entries):")
display(df[['Name', 'Age', 'Age_Category']].head())
```

 Data Preprocessing & Transformation:


### One-Hot Encoding Categorical Variables (`pd.get_dummies()`)

```python
import pandas as pd

# Select categorical columns to encode
categorical_cols = ['Preferred Foot', 'Body Type']

# Apply one-hot encoding using pd.get_dummies()
df_encoded = pd.get_dummies(df, columns=categorical_cols, prefix=categorical_cols)

print("Original 'Preferred Foot' and 'Body Type' columns (first 5 entries):")
display(df[['Preferred Foot', 'Body Type']].head())

print("\nDataFrame after one-hot encoding (first 5 entries, showing all new columns):")
display(df_encoded.head())
```

### Data Preprocessing & Transformation

#### Normalize Numerical Columns (Min-Max Scaling)

```python
import numpy as np

# Choose a numerical column for normalization, e.g., 'Overall'
overall_column = df['Overall']

# Perform Min-Max Scaling
min_overall = overall_column.min()
max_overall = overall_column.max()
df['Overall_Normalized'] = (overall_column - min_overall) / (max_overall - min_overall)

print("Original 'Overall' (first 5 entries):")
display(df[['Overall', 'Overall_Normalized']].head())
```

#### Apply Transformations (Log Scaling)

```python
import numpy as np

# Choose a numerical column for log transformation, e.g., 'Wage'
# Convert 'Wage' from string (e.g., '2K') to numeric
df['Wage_Numeric'] = df['Wage'].replace({'': '', 'K': 'e3', 'M': 'e6'}, regex=True).astype(float)

# Apply log transformation (using log1p to handle potential zero values)
df['Wage_Log'] = np.log1p(df['Wage_Numeric'])

print("Original 'Wage' and 'Wage_Log' (first 5 entries):")
display(df[['Wage', 'Wage_Numeric', 'Wage_Log']].head())
```

### Optimize DataFrame Performance by Changing Data Types

```python
print("Memory usage before optimization:")
display(df.info(memory_usage='deep'))

# Convert suitable integer columns to smaller integer types
for col in ['ID', 'Age', 'Overall', 'Potential', 'Special', 'Weak Foot', 'Skill Moves', 'International Reputation']:
    if df[col].dtype == 'int64':
        if df[col].max() < np.iinfo(np.int16).max and df[col].min() > np.iinfo(np.int16).min:
            df[col] = df[col].astype(np.int16)
        elif df[col].max() < np.iinfo(np.int32).max and df[col].min() > np.iinfo(np.int32).min:
            df[col] = df[col].astype(np.int32)

# Convert suitable float columns to smaller float types
for col in df.select_dtypes(include=['float64']).columns:
    if df[col].min() >= np.finfo(np.float32).min and df[col].max() <= np.finfo(np.float32).max:
        df[col] = df[col].astype(np.float32)

# Convert object columns with few unique values to 'category' dtype
for col in df.select_dtypes(include=['object']).columns:
    if df[col].nunique() < len(df) / 2: # Heuristic: if unique values are less than half the total rows
        df[col] = df[col].astype('category')

print("\nMemory usage after optimization:")
display(df.info(memory_usage='deep'))
```

#### Find correlations between numerical features using `df.corr()`

```python
import seaborn as sns

# Select a few numerical columns for correlation analysis
# Replaced 'Attack', 'Defense', 'Speed' with 'Special', 'Weak Foot', and 'Skill Moves' which are present in the FIFA dataset
numerical_cols_for_corr = ['Age', 'Overall', 'Potential', 'Special', 'Weak Foot', 'Skill Moves', 'Wage_Numeric']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols_for_corr].corr()

print("Correlation Matrix:")
display(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Selected Numerical Features')
plt.show()
```
